Titanic Dataset Analysis and Visualization using Python, NumPy, Matplotlib and, Seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/devra/Data Science/Projects/Titanic Dataset/Data/train.csv')

#detail of the first 15 rows of the data set.
data.head(15)
```

```python
#check number of rows and columns
data.shape
```

```python
#check name of the columns
data.columns
```

```python
data.info()
```

```python
# going to delete PassengerId column as it is not going to give much inteligence to our dataset
# axis = 1 means we are going to remove the whole column
data.drop(['PassengerId'], axis=1, inplace=True)
data.columns
```

```python
data.shape
```

```python
# Check average fare paid and average age of the passengers
# As we want to see average fare and average age, we will use 'mean' function for this.
print("Average Fare: ", data['Fare'].mean())
print("Average Age: ", data['Age'].mean())
```

```python
# we are getting result in long desimal numbers so, we will round off it to 2 desimal places.
print("Average Fare: ", round(data['Fare'].mean(),2))
print("Average Age: ", round(data['Age'].mean(),2))
```

```python
# now generate a graph for survived and not survaived passengers.
sns.countplot(data['Survived'])
plt.title("Not Survived and Survived")
plt.xlabel("Survival")
plt.ylabel("Count")
plt.show()
```

```python
# check the nuber of passengers in class one, class two and class tree.

print("Number of passengers in: ")
first_class = (data['Pclass']== 1).sum()
print("First Class: ", first_class)

second_class = (data['Pclass']== 2).sum()
print("Second Class: ", second_class)

third_class = (data['Pclass']== 3).sum()
print("Third Class: ", third_class)
```

```python
# Pie chart of the number of the passengers in different 3 classes.

labels = ['First Class', 'Second Class', 'Third Class']
sizes = [first_class, second_class, third_class]
plt.pie(sizes, labels=labels, autopct = '%1.2f%%') # autopct = '%1.2f%%' will show percentage in the chart.
plt.title("Passenger Class Distribution")
plt.show()
```

```python
# check number of servived and died people per class

# increase the figure size
plt.figure(figsize=(14,6))

sns.countplot(x=data["Pclass"], hue = data["Survived"])
plt.title("Survived and Died with respect to Class")
plt.xlabel("Classes")
plt.ylabel("Count")
plt.show()
```

```python
# Checking distribution of ages of the passengers

data.hist(column='Age', bins=90, figsize=(14, 6))
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# check age distribution of survived and died people.
plt.figure(figsize = (14,6))
axs = sns.kdeplot(data.Age[data.Survived == 0], label="Died")
axs = sns.kdeplot(data.Age[data.Survived == 1], label="Servived")
plt.title("Survival with respect to Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# adding shade in the curves

plt.figure(figsize = (14,6))
axs = sns.kdeplot(data.Age[data.Survived == 0], shade=True, label="Died")
axs = sns.kdeplot(data.Age[data.Survived == 1], label="Servived")
plt.title("Survival with respect to Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Checking distribution of Fare

data.hist(column='Fare', bins=200, figsize=(14, 6))
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Number of people who survived after paying more than USD 500
data.loc[data['Fare']>500]
```

```python
# Another way
survived_500 =((data['Fare']>500) & (data['Survived']==1)).sum()
print("Survived after paying more than USD 500: ", survived_500)
```

```python
# using scatter plot to visualize age with respect to fare.

plt.figure(figsize = (12,8))
plt.scatter(data['Age'], data['Fare'])
plt.title("Age distribution with respect to Fare.")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# Number of men survived and died

male_died = ((data['Sex']=='male') & (data['Survived']==0)).sum()
print("Number of men died: ", male_died)

male_survived = ((data['Sex']=='male') & (data['Survived']==1)).sum()
print("Number of men survived: ", male_survived)
```

```python
# Number of women survived and died

female_died = ((data['Sex']=='female') & (data['Survived']==0)).sum()
print("Number of women died: ", female_died)

female_survived = ((data['Sex']=='female') & (data['Survived']==1)).sum()
print("Number of women survived: ", female_survived)
```

```python
# bar chart of Men VS Women Survived and Died.

plt.figure(figsize = (12,8))
male_data = (male_died, male_survived)
female_data = (female_died, female_survived)
p1 = plt.bar(np.arange(2), (male_data), width = 0.5)
p2 = plt.bar(np.arange(2), (female_data), bottom= male_data, width = 0.5)
plt.xticks(np.arange(2), ['Men', 'Women'])
plt.legend((p1[0], p2[0]),("Dead", "Survived"))
plt.title("Men VS Women Survival")
plt.show()
```

```python
# Check Boarding stations of passengers

sns.countplot(x=data['Embarked'])
plt.title("Bording stations count")
plt.xlabel("Port")
plt.ylabel("Count")
plt.show()
```

```python
# check survivals with respect to boarding ports
# here S = Southampton, C = Cherbourg, Q = Queenstown

sns.countplot(x=data['Embarked'], hue=data['Survived'])
plt.title("Bording stations survival count")
plt.xlabel("Port")
plt.ylabel("Count")
plt.show()
```

```python
# Generate HeatMap of the dataset

plt.figure(figsize = (12,8))
ax= sns.heatmap(data.corr().abs(), annot=True)
plt.show()
```

```python
# In the above heatmap, Sex and Embarked columns are not included because those are categorical 
#values rather than Numerical values.

# Converting categorical values of columns Sex and Embarked to numarical values.

data['Sex'] = data['Sex'].replace('female',0)
data['Sex'] = data['Sex'].replace('male',1)

data['Embarked'] = data['Embarked'].replace('C',1)
data['Embarked'] = data['Embarked'].replace('Q',2)
data['Embarked'] = data['Embarked'].replace('S',3)

plt.figure(figsize = (14,10))
ax= sns.heatmap(data.corr().abs(), annot=True)
plt.show()
```
